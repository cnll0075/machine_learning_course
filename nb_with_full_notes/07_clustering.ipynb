### Perform clustering

```python
from sklearn.cluster import KMeans
```

```python
inertias = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300, n_init=12, random_state=0)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)
# plt.subplots(figsize=(8, 6))
plt.plot(range(1,11),inertias)
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

Write a for loop to calculate the intertia for k = 1 - 10. Plot the relationship between k and inertia and choose the best K (elbow method).

```python
clusterNum = 5
k_means =KMeans(init='k-means++', n_clusters=clusterNum, n_init=10)
k_means.fit(df_scaled)
labels = k_means.labels_
print(labels)
```

```python
df['cluster_id'] = labels
```

```python
plt.subplots(figsize=(16, 12))
plt.scatter(df['alcohol'], df['total sulfur dioxide'], c=labels.astype(np.float), alpha=.5)
plt.xlabel('alcohol', fontsize=18)
plt.ylabel('total sulfur dioxide', fontsize=16)
```
