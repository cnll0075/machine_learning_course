### Perform clustering

```python
model = KMeans(n_clusters=2, init='k-means++', max_iter=200, n_init=12, random_state=0)
```

Write a for loop to calculate the intertia for k = 1 - 10. Plot the relationship between k and inertia and choose the best K (elbow method).

```python
inertias = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=200, n_init=12, random_state=0)
    kmeans.fit(df_scaled)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 11), inertias)
plt.xlabel("Number of clusters")
plt.title("Elbow??")
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=200, n_init=12, random_state=0)
kmeans.fit(df_scaled)
labels = kmeans.labels_
```

```python
len(labels)
```

```python
df['cluter_id'] = labels
```

```python
df.head()
```

```python
plt.subplots(figsize=(16, 12))
plt.scatter(df['alcohol'], df['total sulfur dioxide'], c=labels.astype(np.float), alpha=0.5)
plt.xlabel('alcohol')
plt.ylabel('total sulfur dioxide')
plt.show()
```
